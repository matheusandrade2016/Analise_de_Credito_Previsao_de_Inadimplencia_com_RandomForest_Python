```python
# Importando minha biblioteca 
# sklearn.ensemble -- > Significa que iremos unir varios algoritimos em 1 só pacote 
from sklearn.ensemble import RandomForestClassifier
```

```python
# Importando minha base credit
import pickle
with open('credit.pkl', 'rb') as f:
  x_credit_treinamento,y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
# Vericando o tamanho do treinamendo
x_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
# Verificando o tamanho do teste
x_credit_teste.shape, y_credit_teste.shape
```

```python
# Criando minha variavel de Random Forest
# Paramentos
  # n_estimators = 100 --> Numeros de arvores que ira usar quanto mais alto for mais ele ira treinar
  # critetion='gini' --> Ele vem como padrao porem iremos usa o 'entropy' como as aulas teoricas

# Sendo Assim minha melhor porcentagem foi de 50 arvores de decisão (n_estimators=50)

random_forest_credit = RandomForestClassifier(n_estimators=50, criterion='entropy',random_state= 0)
# Fit --> Treinaar o algoritimo
random_forest_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
# Criando minha Variavel Previsoes do algoritimo 
previsoes = random_forest_credit.predict(x_credit_teste)
previsoes
```

```python
# Comparando com as respostas reais
y_credit_teste
```

```python
# Mostrando meu accuray_score para ver a porcentagem 

# Precisamos sempre ver qual a melhor solução pois o RandomForest ele e um algoritimo pesado sendo
# na minha arvore de decisão ele deu praticamento o mesmo resultado 

from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_credit_teste, previsoes)
```

```python
# Verificando meu matrix_confusion (matriz de confusão)
from yellowbrick.classifier import ConfusionMatrix;
cm = ConfusionMatrix(random_forest_credit)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)

# Sendo assim na Classe 1 clientes que nao vão pagar a Arvore de Decisão foi melhor 
```

```python
# Mostrando meu relatorio de classificação (classification_report)
print(classification_report(y_credit_teste, previsoes))
```
